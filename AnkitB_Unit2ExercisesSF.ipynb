Why do we make guesses? In other words, what is the benefit of trying to predict something we are uncertain about?

We make guesses to try to estimate the probability of a certain event. We need to make predictions because it's sometimes impossible to know something for sure. In these cases, the only option is to make an educated guess.

If no, then briefly justify why we need assumptions to make predictions.

No, you need assumptions to make any logical guess. In order to make a prediction, there needs to be some logic behind it, but this assumes that the factors that effect it remain constant. **Task3**:

Should we use all the available information we have to make a guess/prediction? Justify your answer.

We don't need to use all available information when making predictions. In certain circumstances, outstanding factors influence data, and it's best to remove those anomalies. Using all information without fail opens statistical models up to distortion from human error. For example, a lab may remove values that vary more than the standard deviation to normalize the data. This would help the results become more realistic/reliable.

What is a prior? How are priors related to
- context?

- predictions?

A prior is a prediction of the probability using the information on hand before knowledge of some data. In order to make a prediction, you have to assume that circumstances are cross applicable or use uncertain data. **Task5**:

What is a likelihood? How are likelihoods related to:

- context?
- assumptions?
- predictions?

The likelihood is the probability of reaching the true probability with your given assumptions in the prior. The likelihood makes an assumption that your premises for the prior are true and that the true probability is accurate. In predictions, it's used to evaluate whether the data matches a given guess.

**Task6**:

What is a posterior? How are posteriors related to:

- context?
- assumptions?
- predictions?

The posterior is an estimate of the probability taking new information in consideration by using the Bayes theorem to update the prior. In the Webanyama example, it would be the prediction with the actual NBA information. You have to assume that your calculations of the prior and likelihood are correct, and that past circumstances can be equally applied to the new situation.
